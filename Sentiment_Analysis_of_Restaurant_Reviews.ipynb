```python
#import dataset
dataset = pd.read_csv('Restaurant_Reviews.xlsx.tsv', delimiter= '\t')
dataset.head()
```

**Step2 :** Text Cleaning or preprocessing

Remove Punctuations, Numbers: Punctuations, numbers does not help much in processing the given text if included, they will just increase the size of bag of words that we will create as last step and decrease the effciency of algorithm
Stemming: Take roots of the word
Convert every word into a lowercase: such as it's useless to have words that are in a different case (e.g., "good" and "GOOD').

```python
from nltk.corpus.reader import reviews
from nltk import corpus
#library to clean data
import re

#natural language tool kit
import nltk
nltk.download('stopwords')

#to remove stopwords
from nltk.corpus import stopwords

#for stemming propose
from nltk.stem.porter    import PorterStemmer

#initialize empty arry
#to append clean text
corpus = []

#1000 (reviews) rows to clean
for i in range(0,1000):

  #column : "reviews" ,rows ith
  Review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])

  #convert all cases to lower case
  Review = Review.lower()

  #split to arry(defult delimiter is "")
  Review = Review.split()

  #creating portrrstemmer  object to
  #to create back into a string
  ps = PorterStemmer()

  #loop for stemming each word
  #in string array at ith row
  Review = [ps.stem(word) for word in Review
            if not word in set(stopwords.words('english'))]
  #rejoin all string array eliment
  # to creat back into a string
  Review = ' '.join(Review)

  #append each string to create
  # array of clean text
  corpus.append(Review)
```

**Step 3:** Tokenization This involves splitting words and sentences from the text's body.

**Step 4:** The bag of words using a sparse matrix

take all the different words of reviews in the dataset with out reof words
One column for each word , therefore there going to be many columns. Rows are reviews

```python
#creating the bag of words model
from sklearn.feature_extraction.text   import CountVectorizer

# to extract max 1500 feature
# max feature is atribute to
# experiment with to get better result
cv = CountVectorizer(max_features= 1500)

# x contents corpus(dependant variable)
x = cv.fit_transform(corpus).toarray()

#y content answers if reviews
# is positive or negative
y = dataset.iloc[:,1].values
```

**Step 5:**    Separating Corpus in the Test and Training set. For this, we need class train_test_split from sklearn.cross_validation. Split can be constructed 70/30, 70/30, 80/20 or 85/15, or 75/25. In this case, I've chosen 75/25 through "test_size." The words bag and y can be 0 and 1. (positive or negative).

```python
#splitting dataset into
#the training set and test set
from sklearn.model_selection        import train_test_split

# experiment with "test_size"
# to get better result
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

**Step 6:** Use a predictive model (here, we will use RandomForest Classifier

As RandomForest Classifier is an ensemble model (that means it is made of many trees) from sklearn.ensemble library we will import RandomForestClassifier module.

With 501 trees, or "n_estimators" and criterion as "entropy."

```python
#fitting random forest classification
# to the training set
from sklearn.ensemble             import RandomForestClassifier

# n estimators can be said as number of
# trees , experiment with n_estimators
# to get better result

model = RandomForestClassifier(n_estimators= 501,
                               criterion= 'entropy')
model.fit(x_train, y_train)
```

**Step 7:** Predicting End Results using the .predict() technique using the attribute X_test.

```python
# predictindg the test set result
y_pred = model.predict(x_test)

y_pred

acc = round(model.score(x_test, y_test)*100,2)
print(str(acc)+'%')
```
